```python
# 2070. 큰 놈, 작은 놈, 같은 놈
T = int(input())
for i in range(T):
    a, b = map(int, input().split())
    if a > b:
        print('#' + str(i+1) + ' ' + '>')
    elif a == b:
        print('#' + str(i+1) + ' ' + '=')
    elif a < b:
        print('#' + str(i+1) + ' ' + '<')
```

```python
# 2050. 알파벳을 숫자로 변환
S = str(input())
Alpha = {'A':1,'B':2,'C':3,'D':4,'E':5,'F':6,'G':7,'H':8,'I':9,'J':10,'K':11,'L':12,'M':13,'N':14,'O':15,'P':16,'Q':17,'R':18,'S':19,'T':20,'U':21,'V':22,'W':23,'X':24,'Y':25,'Z':26}
for  i in range(len(S)):
    if i != len(S)-1:
        print(Alpha[S[i]], end = ' ')
    elif i == len(S)-1:
        print(Alpha[S[i]])
```

```python
# 2063. 중간값 찾기
n = int(input())
nL = list(input().split())
for i in range(n):
  nL[i] = int(nL[i])
nL = sorted(nL)
print(nL[n//2])
```

```python
# 2019. 더블더블
n = int(input())
t = 1
for i in range(1,n+2):
  if i == 1:
    print(t, end = ' ')
  elif i > 1 and i != n+1:
    t *= 2
    print(t, end = ' ')
  elif i == n+1:
    t *= 2
    print(t)
```

```python
# 1545. 거꾸로 출력해 보아요
n = int(input())
for i in range(n,-1,-1):
    if i != 0:
        print(i, end = ' ')
    elif i == 0:
        print(i)
```

```python
# 2072. 홀수만 더하기
T = int(input())
for i in range(T):
    a = list(input().split())
    asum = []
    for k in range(10):
        a[k] = int(a[k])
        if a[k] % 2 != 0:
            asum.append(a[k])
    print('#' + str(i+1) + ' ' + str(sum(asum)))
```

```python
# 2071. 평균값 구하기
T = int(input())
for i in range(T):
    a = list(input().split())
    for k in range(10):
        a[k] = int(a[k])
    print('#' + str(i+1) + ' '+ str(round(sum(a)/10)))
```

```python
# 2056. 연월일 달력
T = int(input())
M = ['01','03','05','07','08','10','12']
M2 = ['02']
M3 = ['04','06','09','11']
for i in range(T):
    a = str(input())
    if a[0:4] != '0000':
        if a[4:6] in M:
            if int(a[6:8]) in range(1,32):
                print('#' + str(i+1) + ' ' + a[0:4] +'/'+a[4:6]+'/'+a[6:8])
            else:
                print('#' + str(i+1) + ' ' + str(-1))
        elif a[4:6] in M2:
            if int(a[6:8]) in range(1,29):
                print('#' + str(i+1) + ' ' + a[0:4] +'/'+a[4:6]+'/'+a[6:8])
            else:
                print('#' + str(i+1) + ' ' + str(-1))
        elif a[4:6] in M3:
            if int(a[6:8]) in range(1,31):
                print('#' + str(i+1) + ' ' + a[0:4] +'/'+a[4:6]+'/'+a[6:8])
            else:
                print('#' + str(i+1) + ' ' + str(-1))
        else:
            print('#' + str(i+1) + ' ' + str(-1))
    else:
        print('#' + str(i+1) + ' ' + str(-1))
                
```
